# Simple Linear Regression using Pytorch

```python
import torch
import torch.nn as nn
```

```python
from sklearn.datasets import make_regression
```

```python
data=make_regression(n_samples=2000,
                     n_features=10,
                     n_informative=8,
                     n_targets=1,
                     random_state=1)
print(data[0].shape)
print(data[1].shape)
```

```python
x=torch.from_numpy(data[0]).float()
y=torch.from_numpy(data[1].reshape(-1,1)).float()
```

```python
layer1=nn.Linear(10,1)
```

```python
loss_fn=nn.MSELoss()#Mean-Square Error
optimizer=torch.optim.SGD(layer1.parameters(),lr=0.01)#Stochastic Gradient Descent
```

```python
epochs=100
losses=[]
for epoch in range(epochs):
  y_pred=layer1(x)
  loss=loss_fn(y_pred,y)
  losses.append(loss.detach().numpy())
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.show()
```
